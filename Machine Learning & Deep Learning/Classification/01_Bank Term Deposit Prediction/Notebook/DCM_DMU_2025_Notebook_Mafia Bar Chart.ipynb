# Feature Extraction

```python
#mengubah value 999 pada fitur 'hari_sejak_kontak_sebelumnya' agar tidak mengganggu distribusi data
data['status_dihubungi'] = data['hari_sejak_kontak_sebelumnya'].apply(
    lambda x: 'belum' if x == 999 else 'pernah'
)

cols = list(data.columns)
pos = cols.index('hari_sejak_kontak_sebelumnya')
#memindahkan kolom baru ke posisi sebelum fitur 'hari_sejak_kontak_sebelumnya'
cols.insert(pos, cols.pop(cols.index('status_dihubungi')))

#menyusun ulang DataFrame sesuai urutan kolom baru
data = data[cols]
```

```python
encode_jenis_kontak = {
    'telephone': 0,
    'cellular': 1
}
df['jenis_kontak'] = df['jenis_kontak'].replace(encode_jenis_kontak)
data['jenis_kontak_x_jumlah_kontak'] = df['jenis_kontak'] * data['jumlah_kontak_kampanye_ini']
```

```python
data['usia_x_tingkat_variasi'] = data['usia'] * data['tingkat_variasi_pekerjaan']
data['suku_bunga_x_jumlah_pekerja'] = data['suku_bunga_euribor_3bln'] * data['jumlah_pekerja']
```

# Testing Model

```python
import optuna
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import roc_auc_score
import numpy as np

# Menentukan fitur dan target
X = data.drop(columns=['berlangganan_deposito', 'pinjaman_rumah', 'pinjaman_pribadi', 'suku_bunga_euribor_3bln', 'pulau', 'hari_sejak_kontak_sebelumnya', 'jumlah_kontak_sebelumnya'])  
y = data['berlangganan_deposito']

# Kolom kategorikal
cat_features = ['status_dihubungi', 'pekerjaan', 'status_perkawinan', 'pendidikan', 'gagal_bayar_sebelumnya', 
                'jenis_kontak', 'bulan_kontak_terakhir', 'hari_kontak_terakhir', 'hasil_kampanye_sebelumnya']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=46)

# Fungsi objective untuk Optuna
def objective(trial):
    params = {
        "iterations": trial.suggest_int("iterations", 100, 500),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.1),
        "depth": trial.suggest_int("depth", 4, 10),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1.0, 5.0),
        "border_count": trial.suggest_int("border_count", 32, 128),
        "random_state": 46,
        "verbose": 0,
        "cat_features": cat_features,
        "auto_class_weights": "Balanced"
    }

    model = CatBoostClassifier(**params)
    cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=46)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc', n_jobs=-1)
    return np.mean(scores)

# Jalankan Optuna
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10, n_jobs=-1)

# Tampilkan hasil terbaik
print("Best Trial:")
print("  Value (ROC AUC):", study.best_trial.value)
print("  Params:", study.best_trial.params)

# Latih ulang model dengan parameter terbaik
best_params = study.best_trial.params
best_params.update({
    "random_state": 46,
    "verbose": 0,
    "cat_features": cat_features,
    "auto_class_weights": "Balanced"
})

final_model = CatBoostClassifier(**best_params)
final_model.fit(X_train, y_train)

# Evaluasi pada test set
y_pred_proba = final_model.predict_proba(X_test)[:, 1]
print("ROC AUC (Test):", roc_auc_score(y_test, y_pred_proba))
```
